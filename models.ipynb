This notebook defines all data models used in the Cyber Cafe Billing System.

##  Cell 1 — Import Required Libraries

```python
from dataclasses import dataclass, field
from datetime import datetime

print('Libraries imported successfully!')
```

##  Cell 2 — Member Dataclass
Stores information about registered cafe members who get discounts.

```python
@dataclass
class Member:
    name: str
    phone: str
    member_id: int = None
    discount_percent: float = 30.0
    join_date: str = field(default_factory=lambda: datetime.now().strftime('%Y-%m-%d'))

# Test
m = Member(name='Ali Khan', phone='03001234567')
print(m)
```

##  Cell 3 — Session Dataclass
Tracks each computer session — start time, end time, which computer, which customer.

```python
@dataclass
class Session:
    computer_no: int
    customer_name: str
    session_id: int = None
    member_id: int = None
    start_time: str = field(default_factory=lambda: datetime.now().strftime('%Y-%m-%d %H:%M:%S'))
    end_time: str = None
    total_bill: float = 0.0
    is_active: bool = True

# Test
s = Session(computer_no=3, customer_name='Sara Ahmed')
print(s)
```

##  Cell 4 — Bill Dataclass
Represents the final bill generated after a session ends.

```python
@dataclass
class Bill:
    session_id: int
    customer_name: str
    computer_no: int
    start_time: str
    end_time: str
    duration_minutes: float
    rate_per_hour: float
    discount: float
    total_amount: float

# Test
b = Bill(
    session_id=1,
    customer_name='Sara Ahmed',
    computer_no=3,
    start_time='2026-04-28 10:00:00',
    end_time='2026-04-28 11:30:00',
    duration_minutes=90,
    rate_per_hour=30.0,
    discount=0.0,
    total_amount=45.0
)
print(b)
```

##  Cell 5 — Summary
All 3 models are ready to use in the rest of the project.

```python
print('=== Cyber Cafe Models Summary ===')
print(f'Member  : {Member.__dataclass_fields__.keys()}')
print(f'Session : {Session.__dataclass_fields__.keys()}')
print(f'Bill    : {Bill.__dataclass_fields__.keys()}')
```
